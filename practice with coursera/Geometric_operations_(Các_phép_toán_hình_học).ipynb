```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open("3.jpg")
```

```python
image.show(title = "a happy girl")
```

```python
plt.imshow(image)
```

# Thay i kch thc hnh nh (scale)

```python
width, height = image.size
```

```python
print(f"Chiu rng l: {width} v chiu di l {height}")
```

Ly chiu rng v chiu cao bng hm **.size**

```python
new_width = width * 2 # tng chiu rng ln gp 2 ln
new_height = height   # vn gi nguyn kch thc ca chiu cao
```

```python
new_image = image.resize((new_width, new_height))
```

Thay i hnh nh vi cc tham s kch thc mi new_width, new_height bng hm **.resize()**

```python
plt.imshow(image) # show ra nh gc
```

```python
plt.imshow(new_image) # show ra nh sau khi  scale theo chiu rng (width)
```

```python
new_width_2 = width   # gi nguyn chiu rng
new_height_2 = height * 2 # tng chiu di ln gp i
```

```python
new_image_2 = image.resize((new_width_2, new_height_2))
```

```python
plt.imshow(new_image_2) # show nh  scale theo chiu cao (height)
```

```python
new_width_3 = width * 2    # scale c kch thc chiu rng v chiu cao
new_height_3 = height * 2
```

```python
new_image_3 = image.resize((new_width_3, new_height_3))
```

```python
plt.imshow(new_image_3) # show nh  scale theo c chiu rng v chiu cao
```

```python
new_width_4 = width // 2 # scale chiu rng nh xung cn phn na
new_height_4 = height // 2 # scale chiu cao nh xung cn phn na
```

```python
new_image_4 = image.resize((new_width_4, new_height_4))
```

```python
plt.imshow(new_image_4) # hin th nh sau khi  co li chiu rng v chiu cao u gim xung cn mt na
```

# Xoay nh

```python
theta = 45
```

khi to tham s theta = 45   xoay nh

```python
new_image_5 = image.rotate(theta)
```

thc hin xoay nh vi hm **rotate** vi tham s u vo l gc xoay **theta**

```python
plt.imshow(new_image_5) # hin th nh sau khi xoay 1 gc 45 
```

```python
new_image_6 = image.rotate(90) # th xoay nh vi gc 90 
```

```python
plt.imshow(new_image_6) # hin th nh sau khi  xoay 1 gc 90 
```

```python
new_image_7 = image.rotate(180) # xoay nh vi gc 180 
```

```python
plt.imshow(new_image_7) # hin th nh vi gc xoay l 180 
```
